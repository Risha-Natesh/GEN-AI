```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Define the model name
model_name = 'distilbert-base-uncased-finetuned-sst-2-english'

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
print(f"Tokenizer '{model_name}' loaded successfully.")

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained(model_name)
print(f"Model '{model_name}' loaded successfully.")
```

```python
example_texts = [
    "I love using Hugging Face Transformers!",
    "This movie was absolutely terrible and a waste of time.",
    "The weather today is quite pleasant.",
    "I'm not sure how I feel about this.",
    "What a fantastic experience!"
]
print(f"Prepared {len(example_texts)} example texts for sentiment analysis.")
```

```python
inputs = tokenizer(example_texts, padding=True, truncation=True, return_tensors='pt')
print("Example texts tokenized successfully.")
print(f"Tokenized input shape: {inputs['input_ids'].shape}")
```

```python
import torch

# Pass tokenized inputs through the model
with torch.no_grad(): # Disable gradient calculation for inference
    outputs = model(**inputs)
    logits = outputs.logits

print("Inference performed successfully.")
print(f"Logits shape: {logits.shape}")
```

```python
import torch.nn.functional as F

# 1. Apply softmax to get probabilities
probabilities = F.softmax(logits, dim=1)

# 2. Determine the predicted class ID (0 for negative, 1 for positive)
predicted_class_ids = torch.argmax(probabilities, dim=1)

# 3. Map class IDs to meaningful labels
label_map = {0: 'NEGATIVE', 1: 'POSITIVE'}
predicted_labels = [label_map[id.item()] for id in predicted_class_ids]

# 4. Print each original example text along with its predicted sentiment label and confidence score
print("Sentiment Analysis Results:")
print("----------------------------")
for i, text in enumerate(example_texts):
    label = predicted_labels[i]
    confidence = probabilities[i][predicted_class_ids[i]].item()
    print(f"Text: '{text}'")
    print(f"Predicted Sentiment: {label} (Confidence: {confidence:.4f})")
    print("----------------------------")
```
