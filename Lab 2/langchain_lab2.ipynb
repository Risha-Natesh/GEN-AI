```python
import os
from google.colab import userdata

# Set the GROQ_API_KEY as an environment variable
os.environ["GROQ_API_KEY"] = userdata.get('GROQ_API_KEY')

print("Groq API Key loaded and set.")
```

```python
from langchain_groq import ChatGroq
from langchain.prompts import PromptTemplate

# Initialize the LLM with Groq
# You can choose other Groq models like 'mixtral-8x7b-32768', 'llama3-70b-8192', etc.
# Using 'llama3-8b-8192' as a common choice, or you can use 'gemma-7b-it'
llm = ChatGroq(model="llama3-8b-8192")

# Define a simple prompt template
prompt = PromptTemplate.from_template("{question}")

# Create a runnable chain using LangChain Expression Language (LCEL)
chain = prompt | llm

# Ask a question
question = "What is the capital of France?"
response = chain.invoke({"question": question})

print(f"Question: {question}")
print(f"Answer: {response.content}")
```
