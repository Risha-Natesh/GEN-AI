### Problem Definition

**Problem:**

On Thursday, they bake another 100 cookies, but accidentally burn 20% of the cookies made on Thursday. How many unburnt cookies does the baker have at the end of Thursday?

### Crafted Prompt

```text
Problem:
A baker makes 250 cookies on Monday. On Tuesday, they make twice the number of cookies made on Monday. On Wednesday, they sell 75% of the total cookies made on Monday and Tuesday. On Thursday, they bake another 100 cookies, but accidentally burn 20% of the cookies made on Thursday. How many unburnt cookies does the baker have at the end of Thursday?

Instructions:
1. Think step by step to solve the problem. Show all intermediate calculations.
2. State your initial solution.
3. Verify your solution by re-examining each step and calculation. Explain your verification process and identify any potential errors or confirm the correctness of your steps.
4. Provide the final, verified answer.
```

## Illustrate Question Refinement

### Subtask:
Provide a scenario demonstrating an initial broad question, a basic LLM response, and then a refined question to get more specific information.


### Scenario: Planning a Trip to Japan

Imagine you are planning your first trip to Japan and are looking for advice on popular destinations and activities.

### Initial Broad Question:
"What are the best places to visit in Japan?"

### Example Basic LLM Response:
"Japan offers a rich blend of traditional culture and modern attractions. Popular destinations include Tokyo (for its vibrant city life, shopping, and dining), Kyoto (known for its temples, gardens, and geishas), and Osaka (famous for its street food and nightlife). Other notable places are Hiroshima, Nara, and Hokkaido for natural beauty."

### Refined Question:
"Given that I'm interested in traditional Japanese culture, ancient temples, and serene gardens, which specific areas within Kyoto would you recommend, and what are some lesser-known cultural experiences or temples I should seek out to avoid large crowds?"

### Example Refined LLM Response:
"For traditional Japanese culture, ancient temples, and serene gardens, Kyoto is indeed an excellent choice. To avoid large crowds while still experiencing the city's charm, consider exploring these areas and experiences:

*   **Arashiyama Bamboo Grove (early morning/late afternoon):** While popular, visiting right at sunrise or an hour before sunset can offer a more peaceful experience. Afterward, explore the quiet **Okochi Sanso Villa**, a stunning traditional Japanese garden and former home of a silent film actor, which often has fewer visitors than other sites.
*   **Fushimi Inari-taisha (upper trails):** The iconic torii gates can be crowded at the bottom, but the further you walk up the mountain, the fewer people you'll encounter, offering a more spiritual and solitary experience.
*   **Northern Kyoto (Kinkaku-ji and Ryoan-ji alternatives):** Instead of just the famous Kinkaku-ji (Golden Pavilion), visit **Daitoku-ji Temple Complex**. It hosts several sub-temples with beautiful Zen gardens and often much fewer tourists. Specifically, **Daisen-in** and **Koto-in** offer profound cultural experiences and stunning gardens.
*   **Higashiyama District (side streets and early visits):** Beyond the main tourist path, wander through the narrow streets early in the morning to find traditional machiya houses, small craft shops, and local tea houses. Visit **Kodai-ji Temple** for its beautiful rock garden and bamboo grove, which is less frequented than Kiyomizu-dera.
*   **Philosopher's Path (off-season or early morning):** This scenic path is beautiful, especially during cherry blossom season, but can be packed. In quieter months or early mornings, it's a tranquil walk between temples.

For a truly unique cultural experience, consider a traditional tea ceremony in a less commercialized setting, or look for local artisan workshops where you can try your hand at crafts like pottery or dyeing. Also, exploring the smaller, local shrines and temples scattered throughout residential areas can offer a glimpse into everyday Japanese spiritual life away from the main tourist hubs."

## Demonstrate Chain of Thought

### Subtask:
Create a multi-step reasoning problem and a prompt that explicitly asks for a step-by-step thought process, followed by an example response.


### 1. Multi-Step Reasoning Problem

**Problem:**

Imagine you are planning a small event. You need to order pizzas from three different pizzerias: 'Pizza Palace', 'Doughnut & Pizza', and 'Crusty Creations'. Each pizzeria offers a different discount based on the number of pizzas ordered.

*   **Pizza Palace:** Offers a 10% discount on the total bill if you order 5 or more pizzas. Each pizza costs $15.
*   **Doughnut & Pizza:** Offers $2 off each pizza if you order 3 or more. Each pizza costs $18.
*   **Crusty Creations:** Offers a fixed $5 discount on the total bill, regardless of the quantity. Each pizza costs $16.

You need to order a total of 12 pizzas for the event. You want to allocate the orders to these pizzerias such that you get at least 3 pizzas from each pizzeria, and you want to minimize the total cost. How many pizzas should you order from each pizzeria to achieve the lowest possible total cost, and what would that total cost be?

### 2. Prompt for Chain of Thought

**Prompt:**

Given the pizza ordering problem, please outline your thought process step by step to determine the optimal allocation of pizzas to minimize cost. Show all your calculations and reasoning for each step. After your detailed step-by-step reasoning, provide the final number of pizzas to order from each pizzeria and the total minimum cost. Think step by step.

### 3. Example LLM Response (Chain of Thought)

**Example Response:**

Okay, I will break down the pizza ordering problem step by step to find the optimal allocation that minimizes the total cost, ensuring at least 3 pizzas from each pizzeria and a total of 12 pizzas.

**Step 1: Understand the Constraints and Requirements**

*   Total pizzas needed: 12
*   Minimum pizzas per pizzeria: 3
*   Pizzerias:
    *   **Pizza Palace (PP):** $15/pizza, 10% discount for >= 5 pizzas.
    *   **Doughnut & Pizza (DP):** $18/pizza, $2 off each pizza for >= 3 pizzas.
    *   **Crusty Creations (CC):** $16/pizza, fixed $5 discount.

**Step 2: Calculate the effective cost per pizza/order for each pizzeria under their discount conditions.**

To effectively compare costs, I need to understand the per-pizza cost given the discounts and minimum order requirements.

*   **Pizza Palace (PP):**
    *   Base cost per pizza: $15
    *   Discount condition: 10% off total for >= 5 pizzas.
    *   If 3 pizzas: 3 * $15 = $45
    *   If 5 pizzas: 5 * $15 = $75. With 10% discount: $75 * 0.90 = $67.50. Effective cost per pizza: $67.50 / 5 = $13.50.

*   **Doughnut & Pizza (DP):**
    *   Base cost per pizza: $18
    *   Discount condition: $2 off each pizza for >= 3 pizzas.
    *   Effective cost per pizza when ordering >= 3: $18 - $2 = $16.
    *   If 3 pizzas: 3 * $16 = $48.

*   **Crusty Creations (CC):**
    *   Base cost per pizza: $16
    *   Discount condition: fixed $5 off total.
    *   If 3 pizzas: (3 * $16) - $5 = $48 - $5 = $43. Effective cost per pizza: $43 / 3 = ~$14.33.

**Summary of potential costs (per pizza where applicable, or total for minimum 3):**
*   PP: $15/pizza (for < 5), $13.50/pizza (for >= 5)
*   DP: $16/pizza (for >= 3)
*   CC: Variable per-pizza cost due to fixed discount, but cheapest for 3 pizzas is $43.

**Step 3: Fulfill the minimum order requirement for each pizzeria first.**

Each pizzeria must have at least 3 pizzas. So, I will allocate 3 pizzas to each initially.

*   Pizza Palace (PP): 3 pizzas * $15 = $45
*   Doughnut & Pizza (DP): 3 pizzas * $16 = $48 (discount applied)
*   Crusty Creations (CC): 3 pizzas * $16 - $5 = $43 (discount applied)

*Current total pizzas allocated: 3 + 3 + 3 = 9 pizzas.*
*Remaining pizzas to allocate: 12 - 9 = 3 pizzas.*
*Current total cost: $45 + $48 + $43 = $136.*

**Step 4: Determine the best place to allocate the remaining 3 pizzas.**

I need to add 3 more pizzas to minimize the total cost. I will evaluate the marginal cost of adding pizzas to each pizzeria.

*   **If I add 3 pizzas to Pizza Palace (PP):**
    *   Current PP pizzas: 3. Adding 3 makes it 6 pizzas.
    *   6 pizzas from PP: 6 * $15 = $90. With 10% discount: $90 * 0.90 = $81.
    *   Cost increase for PP: $81 (new total) - $45 (old total) = $36.
    *   Total cost if adding 3 to PP: $136 + $36 = $172.

*   **If I add 3 pizzas to Doughnut & Pizza (DP):**
    *   Current DP pizzas: 3. Adding 3 makes it 6 pizzas.
    *   6 pizzas from DP: 6 * $16 = $96 (discount already applied per pizza).
    *   Cost increase for DP: $96 (new total) - $48 (old total) = $48.
    *   Total cost if adding 3 to DP: $136 + $48 = $184.

*   **If I add 3 pizzas to Crusty Creations (CC):**
    *   Current CC pizzas: 3. Adding 3 makes it 6 pizzas.
    *   6 pizzas from CC: (6 * $16) - $5 = $96 - $5 = $91.
    *   Cost increase for CC: $91 (new total) - $43 (old total) = $48.
    *   Total cost if adding 3 to CC: $136 + $48 = $184.

Comparing the cost increases:
*   Adding 3 to PP: $36
*   Adding 3 to DP: $48
*   Adding 3 to CC: $48

Adding the remaining 3 pizzas to Pizza Palace yields the lowest additional cost.

**Step 5: Final Allocation and Total Cost Calculation.**

*   **Pizza Palace (PP):** 3 (initial) + 3 (remaining) = 6 pizzas.
    *   Cost for PP: 6 * $15 = $90. With 10% discount: $90 * 0.90 = $81.
*   **Doughnut & Pizza (DP):** 3 pizzas.
    *   Cost for DP: 3 * ($18 - $2) = 3 * $16 = $48.
*   **Crusty Creations (CC):** 3 pizzas.
    *   Cost for CC: (3 * $16) - $5 = $48 - $5 = $43.

Total pizzas: 6 + 3 + 3 = 12 pizzas.
Total cost: $81 + $48 + $43 = $172.

**Final Answer:**

To achieve the lowest possible total cost, you should order:
*   **Pizza Palace:** 6 pizzas
*   **Doughnut & Pizza:** 3 pizzas
*   **Crusty Creations:** 3 pizzas

The total minimum cost would be **$172**.

## Explain RGC Approach

### Subtask:
Explain how to structure a series of prompts to practice the Retrieve, Generate, Critique/Consolidate (RGC) technique.


The RGC (Retrieve, Generate, Critique/Consolidate) approach is a multi-turn prompting technique designed for complex tasks, aiming to enhance the quality and accuracy of AI-generated responses. It breaks down a complex problem into sequential, manageable steps, allowing the Language Model (LLM) to iteratively refine its output.

### 1. Retrieve Phase

**Purpose:** The primary goal of the 'Retrieve' phase is to gather all necessary and relevant information before attempting to generate a final response. This ensures the LLM has a solid factual basis, reducing the likelihood of hallucinations or inaccurate outputs. It mimics the research step a human would undertake.

**Example Prompts for the LLM to Retrieve Information:**

*   "Find and summarize key information about [Topic X], focusing on [specific aspects]."
*   "Extract all relevant data points regarding [Entity Y] from the provided text/document/URL. Present them as a bulleted list."
*   "Research the main arguments for and against [Concept Z] and present a concise summary of each side."
*   "Identify and list the top three factors contributing to [Phenomenon A] based on available knowledge."
*   "Based on the following text, identify all mentions of [Keyword/Phrase] and provide their context."

### 2. Generate Phase

**Purpose:** In the 'Generate' phase, the LLM uses the information retrieved in the previous step to create initial content. This is where the core output is formed, leveraging the pre-researched facts and insights. The quality of this generation heavily depends on the thoroughness of the retrieval phase.

**Example Prompts for the LLM to Generate Content:**

*   "Using the information retrieved in the previous step about [Topic X], draft an initial report (or essay, email, summary, etc.) on [Specific Aspect]."
*   "Based on the data points extracted for [Entity Y], compose a comprehensive description of its characteristics and functions."
*   "Synthesize the arguments for and against [Concept Z] into a balanced discussion. Ensure all points are supported by the summarized facts."
*   "Formulate a paragraph explaining [Phenomenon A], incorporating the identified factors and their relationships."
*   "Write a short article about [Event B], integrating the contextual information gathered about [Keyword/Phrase]."

### 3. Critique/Consolidate Phase

**Purpose:** The 'Critique/Consolidate' phase is crucial for refining, evaluating, combining, or correcting the generated content. This step acts as a self-correction mechanism, allowing the LLM to review its own output, identify weaknesses, and improve the overall quality and accuracy of the final response. It ensures that the generated content meets specific criteria and is free from errors or inconsistencies.

**Example Prompts for the LLM to Critique/Consolidate Content:**

*   "Review the generated report for accuracy, coherence, and completeness. Suggest specific improvements and correct any errors."
*   "Compare the two generated options (e.g., from different 'Generate' turns) and consolidate them into a single, optimized version. Highlight the changes made."
*   "Identify any unsupported claims or logical inconsistencies in the generated text. Explain why they are unsupported or inconsistent and propose corrections."
*   "Evaluate the generated content against the following criteria: [List specific criteria like 'clarity', 'conciseness', 'adherence to prompt', 'factual accuracy']. Provide a score for each and suggest improvements where needed."
*   "Refine the tone and style of the generated article to be more suitable for a [specific audience, e.g., 'technical audience', 'general public', 'formal business setting']."

### Zero-shot Prompting

Zero-shot prompting is a technique where the Large Language Model (LLM) is given a task without any examples of how to perform it. The model relies solely on its pre-existing knowledge to generate a response.

**Example Prompt:**

```
Classify the sentiment of the following sentence as positive, negative, or neutral:
"The new software update significantly improved performance."
```

**Example LLM Response:**

```
Sentiment: Positive
```

```markdown
### One-shot Prompting

One-shot prompting provides the LLM with a single input-output example to guide its response for a new, similar input. This helps the model understand the desired format, style, or task behavior more clearly than zero-shot.

**Example Prompt:**

```
Translate the following English sentence to French:
English: Hello
French: Bonjour

Translate the following English sentence to French:
English: Goodbye
French:
```

**Example LLM Response:**

```
French: Au revoir
```


### Few-shot Prompting

Few-shot prompting extends one-shot prompting by providing the LLM with several (e.g., 2-3) input-output examples. This approach gives the model a more robust understanding of the desired task, format, or style, leading to more accurate and consistent responses for new inputs.

**Example Prompt:**

```
Extract the key entity and its type from the following texts:
Text: Apple Inc. was founded by Steve Jobs.
Entity: Apple Inc.
Type: Company

Text: The Eiffel Tower is in Paris.
Entity: Eiffel Tower
Type: Landmark

Text: Leonardo da Vinci painted the Mona Lisa.
Entity: Leonardo da Vinci
Type: Person

Text: Google acquired YouTube in 2006.
Entity:
Type:
```

**Example LLM Response:**

```
Entity: Google
Type: Company
